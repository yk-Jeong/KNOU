```python
# 판다스 라이브러리 로드 

import pandas as pd
```

```python
# 학생들의 데이터를 저장할 빈 배열 생성 
student = []

# 0이 들어올 때까지 반복 입력
s = 1
while s != ['0']:
  s = list(map(str, input("학생의 이름과 소속, 국어, 영어, 수학의 점수를 입력하세요.(끝이면 0을 입력하세요)").split()))
  student.append(s)

# 마지막에 들어오는 0은 우리가 원하는 데이터가 아니므로 잘라냄 
student = student[:-1]
print(student)
```

```python
# 입력한 테스트에서 숫자 항목이 문자열 변수로 입력되었으므로 형변환함 
# 이때, 숫자와 숫자 사이를 콤마로 구분한 경우 콤마를 먼저 삭제해 주어야 함 

for std in student:
  for i in range(2, len(std)):
    if ',' in std[i]:
      std[i] = std[i].replace(',', '')
    std[i] = int(std[i])

# 잘 변환되었는지 확인
student
```

```python
# 데이터를 핸들링하기 쉽도록 리스트를 dataframe 형태로 변환

df = pd.DataFrame(student, columns = ['이름', '소속', '국어', '영어', '수학'])
df
```

```python
# 학생 목록에서 '컴퓨터과학과'에 재학 중인 학생들의 평균 성적만을 계산 
# numeric_only = True 옵션을 주면 숫자가 아닌 행이나 열을 건너뛰어줌 

mean_com = df[df['소속'] == '컴퓨터과학과'].iloc[:2].mean(numeric_only=True)
mean_com

# 위 결괏값에서 각 과목을 각각의 변수에 할당
mean_com_kor = mean_com[0]
mean_com_eng = mean_com[1]
mean_com_math = mean_com[2]
```

```python
print(f'컴퓨터과학과 학생의 평균 점수는 국어 {mean_com_kor}, 영어 {mean_com_eng}, 수학 {mean_com_math}점입니다.')
```
