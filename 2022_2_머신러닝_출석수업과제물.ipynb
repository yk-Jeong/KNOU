```python
import math
pt1=[2, 2]
def euclidean_distance(pt2 ):
    return math.sqrt( sum( [ ( d1 - d2 )**2 for d1, d2 in zip( pt1, pt2 ) ] ) )
```

```python
distant_list = []
```

```python
for i in range(5):
  d = [A_x[i], A_y[i]]
  distance = euclidean_distance(d)
  distant_list.append(distance)
```

```python
for i in range(5):
  d = [B_x[i], B_y[i]]
  distance = euclidean_distance(d)
  distant_list.append(distance)
```
